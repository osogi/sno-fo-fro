```python
import h2o
from h2o.automl import H2OAutoML

h2o.init()

data_path = 'metrics_table.csv'
data = h2o.import_file(data_path)

print(data.head())
print(data.describe())

target = 'class_label'
features = data.columns[:-1]
train, test = data.split_frame(ratios=[0.8], seed=42)

aml = H2OAutoML(max_models=40, seed=42, balance_classes=True, max_runtime_secs=600)
aml.train(x=features, y=target, training_frame=train)

print(aml.leaderboard)
leader_model = aml.leader

print("Лучшая модель (лидер):", leader_model)

preds = leader_model.predict(test)

print(preds.head())

predictions_df = preds.as_data_frame()

model_path = h2o.save_model(model=leader_model, path="./pretrained", force=True)
print(f"Модель сохранена по пути: {model_path}")
```
